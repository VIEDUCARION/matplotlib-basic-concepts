```python
#STACK BAR CHATS

from matplotlib import pyplot as plt
import numpy as np


years = ["2016","2021","2152","2102","3231"]
cse = [75,66,54,55,45]
it=[36,54,65,49,39]
ece=[58,69,47,49,57]
eee=[35,56,68,78,71]
w = 0.2
plt.figure(figsize = (6,4))

#LOCATIONS
cse_bar = np.arange(len(years))
it_bar = [i+w for i in cse_bar]
ece_bar = [i+w for i in it_bar]
eee_bar = [i+w for i in ece_bar]


plt.bar(cse_bar,cse,width = w,label = "CSE")

plt.bar(it_bar,it,width = w,label = "IT")

plt.bar(ece_bar,ece,width = w,label = "ECE")

plt.bar(eee_bar,eee,width = w,label = "EEE")

plt.xticks(cse_bar+w+w/2,years)


plt.title("College students")
plt.ylabel("Number of students Placed")
plt.xlabel("Year")
plt.ylim(0,90)
plt.legend()
plt.show()
```
